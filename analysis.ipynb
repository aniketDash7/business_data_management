Importing the required dependencies

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

First Look at the data

```python
number_df = pd.read_csv("number.csv - Sheet1.csv")
money_df = pd.read_csv("individual_revenue.csv - Sheet1.csv")
print(number_df.shape)
print(money_df.shape)
```

Getting the basic statistics of the money_df

```python
money_df.describe()
```

Creating the total revenue column by adding up the revenue per day 

```python
money_df['total_revenue_perday'] = money_df.iloc[:, 1:].sum(axis=1)
money_df.head()
```

Establishing fixed cost 

```python
cost_cylinder = 1800/30
cost_chicken = 2100
cost_diesel = 2000/30
cost_staff = 20000/30
total_cost_price_perday = cost_chicken + cost_cylinder + cost_diesel + cost_staff
print("Total Fixed expenditure per day is", total_cost_price_perday)
```

Making a copy of money_df

```python
proposed_money_df = money_df
```

Creating a profit column by deducting the fixed cost from the revenue each day

```python
money_df['profit'] = money_df['total_revenue_perday'] - total_cost_price_perday
money_df.head()
```

Creating a profit-margin column by dividing the profit column by the total revenue column and multiplying by 100 to get a percentage value 

```python
money_df['profit_margin_perday'] = (money_df['profit']/money_df['total_revenue_perday'])*100
money_df.head()
```

Creating a Total units sold/Day column by adding up the rows 

```python
number_df['Total units sold/Day'] = number_df.iloc[:,1:].sum(axis=1)
number_df.head()
```

Creating the feature dataframes

```python
money_df_features = money_df.drop(columns=['Dates','total_revenue_perday','profit','profit_margin_perday'])
number_df_features = number_df.drop(columns=['Dates','Total units sold/Day'])
print(number_df_features.shape)
print(money_df_features.shape)
```

Building a correlation matrix which states the correlation of products on the menu with each other 

```python
correlation_matrix = number_df_features.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.show()
```

SALES VOLUME


```python
number_df['Dates'] = pd.to_datetime(number_df['Dates'])
grouped_by_date = number_df.groupby(number_df['Dates'].dt.month)['Total units sold/Day'].sum()
grouped_by_date_df = pd.DataFrame(grouped_by_date)
grouped_by_date_df.rename(columns={'Total units sold/Day': 'Total Units Sold/Month'}, inplace=True)
grouped_by_date_df.reset_index(inplace=True)
grouped_by_date_df.columns = ['Month', 'Total Units Sold/Month']
grouped_by_date_df
```

```python
print('Sales volume of the month of October 2023 :',grouped_by_date_df['Total Units Sold/Month'][0])
print('Sales volume of the month of November 2023 :',grouped_by_date_df['Total Units Sold/Month'][1])
print('Sales volume of 5 days of December 2023 :',grouped_by_date_df['Total Units Sold/Month'][2])
print('Total Sales Volume :', grouped_by_date_df['Total Units Sold/Month'].sum())

print(f"October 2023 generated {100*(grouped_by_date_df['Total Units Sold/Month'][0]/grouped_by_date_df['Total Units Sold/Month'].sum())}% of Total sales volume")
print(f"November 2023 generated {100*(grouped_by_date_df['Total Units Sold/Month'][1]/grouped_by_date_df['Total Units Sold/Month'].sum())}% of Total sales volume")
print(f"Initial 5 days of December 2023 generated {100*(grouped_by_date_df['Total Units Sold/Month'][2]/grouped_by_date_df['Total Units Sold/Month'].sum())}% of Total sales volume")
```
